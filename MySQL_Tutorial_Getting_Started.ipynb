```
!apt install libmysqlclient-dev > /dev/null
!pip install mysqlclient > /dev/null
import pandas as pd
import MySQLdb
from sqlalchemy import create_engine, text
```

```
connect_string = 'mysql://root@localhost/classicmodels'
engine = create_engine(connect_string)
```

```
#DBConn = MySQLdb.connect(db='praxisDB')
DBConn = engine.connect()
#df_mysql = pd.read_sql('select * from emp2;', con=con_mysql)
df_mysql = pd.read_sql(text('show tables'), con=DBConn)
print ('loaded dataframe from MySQL. records:', len(df_mysql))
DBConn.close()
df_mysql
```

```
#To run any non-SELECT SQL command
def runCMD (DDL):
    #DBConn= MySQLdb.connect(db='praxisDB')
    DBConn = engine.connect()
    #myCursor = DBConn.cursor()
    try:
        #retcode = myCursor.execute(text(DDL))
        retcode = DBConn.execute(text(DDL))
        DBConn.commit()
    except :
        print('Error : return code = ',retcode)
    DBConn.close()

#To run any SELECT SQL command and return data in pandas dataframe
def runSELECT (query):
    #DBConn= MySQLdb.connect(db='praxisDB')
    DBConn = engine.connect()
    df_mysql = pd.read_sql(text(query), con=DBConn)
    #df_mysql = pd.read_sql_query(CMD, con=DBConn)
    DBConn.close()
    return df_mysql
```

##CTE

```
sqlString = "\
SELECT customerName \
 FROM \
    (select customerName, state from  customers \
     where country = 'USA'\
     ) as customers_in_usa \
 WHERE state = 'CA' ORDER BY customerName \
 "
runSELECT(sqlString)
```

```
sqlString = "\
SELECT \
    employees.employeeNumber, firstName, lastName, sales\
 FROM \
    ( \
      SELECT \
        salesRepEmployeeNumber employeeNumber,\
        SUM(quantityOrdered * priceEach) sales\
    FROM\
        orders\
            INNER JOIN\
        orderdetails USING (orderNumber)\
            INNER JOIN\
        customers USING (customerNumber)\
    WHERE\
        YEAR(shippedDate) = 2003\
            AND status = 'Shipped'\
    GROUP BY salesRepEmployeeNumber\
    ORDER BY sales DESC\
    LIMIT 5\
    ) as topsales2003, employees where employees.employeeNumber = topsales2003.employeeNumber"


runSELECT (sqlString)
```

```
sqlString = "\
SELECT  * FROM \
( SELECT  customerName, salesrepName \
    FROM customers, \
        ( SELECT \
        employeeNumber, \
        CONCAT(firstName, ' ', lastName) AS salesrepName \
        FROM   employees \
        WHERE jobTitle = 'Sales Rep' \
        ) as salesrep \
 where employeeNumber = salesrepEmployeeNumber \
) as customer_salesrep \
ORDER BY customerName"

runSELECT (sqlString)
```

##Window Functions in SQL
https://www.geeksforgeeks.org/window-functions-in-sql/

```
#Let us look at payments table
sqlString = 'select * from payments'
runSELECT(sqlString)
```

```
#Basic Group By
sqlString = 'select customerNumber, count(checkNumber) cheques, sum(amount) total from payments group by customerNumber'
runSELECT(sqlString)
```

```
#Pseudo Window Function
sqlString = "\
select payments.customerNumber, checkNumber, paymentDate, amount, cheques, total from payments,\
  ( \
      select customerNumber, count(checkNumber) cheques, sum(amount) total from payments group by customerNumber \
  ) as groupedPayment \
  where payments.customerNumber = groupedPayment.customerNumber \
  "
runSELECT(sqlString)
```
