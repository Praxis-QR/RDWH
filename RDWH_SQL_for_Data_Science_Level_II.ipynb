#Access Functions

```
#To run any non-SELECT SQL command
def runCMD (DDL):
    #DBConn= MySQLdb.connect(hostName,userName,passWord,dbName)   # Remote
    DBConn = MySQLdb.connect(db='praxisDB')                       # Local
    myCursor = DBConn.cursor()
    retcode = myCursor.execute(DDL) 
    print (retcode)
    DBConn.commit()
    DBConn.close()

#To run any SELECT SQL command
def runSELECT (CMD):
    #DBConn= MySQLdb.connect(hostName,userName,passWord,dbName)   # Remote
    DBConn = MySQLdb.connect(db='praxisDB')                       # Local
    df_mysql = pd.read_sql(CMD, con=DBConn)    
    DBConn.close()
    return df_mysql
```

```
runSELECT("select * from EmpDataCols")
```

```
cmd = "\
SELECT \
  empID,\
  GROUP_CONCAT(if(property = 'name', propValue, NULL)) AS 'name',\
  GROUP_CONCAT(if(property = 'address1', propValue, NULL)) AS 'add1',\
  GROUP_CONCAT(if(property = 'address2', propValue, NULL)) AS 'add2',\
  GROUP_CONCAT(if(property = 'education1', propValue, NULL)) AS 'edu1',\
  GROUP_CONCAT(if(property = 'education2', propValue, NULL)) AS 'edu2'\
FROM EmpDataCols \
GROUP BY empID"
runSELECT(cmd)
```

#Generalised Key Value Table

## JSON, Pivot and Group_Concat <br>
1. [Group Concat](https://www.mysqltutorial.org/mysql-group_concat/) is a aggregation function like SUM() or AVG() <br>
2. [JSON](https://www.w3schools.com/whatis/whatis_json.asp) is a very popular key-value data story

### create table

```
runCMD('Drop Table if exists KeyValueTable;')
cmd = 'CREATE TABLE KeyValueTable (            \
  trow int(1) DEFAULT NULL,          \
  tkey varchar(10) DEFAULT NULL,  \
  tvalue varchar(14) DEFAULT NULL  \
)'                              
runCMD(cmd)
```

```
cmd = "INSERT INTO KeyValueTable (trow, tkey, tvalue) VALUES  \
 (1, 'Name', 'Prithwis'),(1, 'Address', 'Calcutta'), (1, 'Role', 'Director'), \
 (2, 'Name', 'Kamlesh'),(2, 'Address', 'Bombay'), (2, 'Role', 'Founder'), \
 (3, 'DeptName', 'Accounts'),(3, 'Location', 'Calcutta'), (3, 'DeptCode', 'D01'), \
 (4, 'DeptName', 'Sales'),(4, 'Location', 'Bombay'), (4, 'DeptCode', 'D21'), \
 (5, 'DeptName', 'Marketing'),(5, 'Location', 'Delhi'), (5, 'DeptCode', 'D99') \
 "                       
runCMD(cmd)
```

```
cmd = "INSERT INTO KeyValueTable (trow, tkey, tvalue) VALUES  \
 (3, 'Mgr', 'pm'), \
 (4, 'Mgr', 'cps'), \
 (5, 'Mgr', 'sdg') "      
runCMD(cmd)
```

```
runSELECT('select * from KeyValueTable')
```

```
cmd = "SELECT \
  trow,\
  GROUP_CONCAT(if(tkey = 'DeptName', tvalue, NULL)) AS 'Department',\
  GROUP_CONCAT(if(tkey = 'Location', tvalue, NULL)) AS 'Location',\
  GROUP_CONCAT(if(tkey = 'DeptCode', tvalue, NULL)) AS 'DepartmentCode' \
FROM KeyValueTable \
GROUP BY trow \
;"      
runSELECT(cmd)
```

```
cmd = "SELECT \
  trow,\
  GROUP_CONCAT(if(tkey = 'DeptName', tvalue, NULL)) AS 'Department',\
  GROUP_CONCAT(if(tkey = 'Location', tvalue, NULL)) AS 'Location',\
  GROUP_CONCAT(if(tkey = 'Mgr', tvalue, NULL)) AS 'Mgr',\
  GROUP_CONCAT(if(tkey = 'DeptCode', tvalue, NULL)) AS 'DepartmentCode' \
FROM KeyValueTable \
GROUP BY trow \
;"      
runSELECT(cmd)
```

```
cmd = "Select t2.Department, t2.Location,t2.DepartmentCode from (\
SELECT \
  trow,\
  GROUP_CONCAT(if(tkey = 'DeptName', tvalue, NULL)) AS 'Department',\
  GROUP_CONCAT(if(tkey = 'Location', tvalue, NULL)) AS 'Location',\
  GROUP_CONCAT(if(tkey = 'DeptCode', tvalue, NULL)) AS 'DepartmentCode' \
FROM KeyValueTable \
GROUP BY trow \
) t2 where Department is NOT NULL \
;"      
runSELECT(cmd)
```

```
cmd = "SELECT \
  trow,\
  GROUP_CONCAT(if(tkey = 'Name', tvalue, NULL)) AS 'Name',\
  GROUP_CONCAT(if(tkey = 'Address', tvalue, NULL)) AS 'Address',\
  GROUP_CONCAT(if(tkey = 'Role', tvalue, NULL)) AS 'Designation' \
FROM KeyValueTable \
GROUP BY trow;"      
runSELECT(cmd)
```

```
cmd = "Select t2.Name, t2.Address,t2.Designation from (\
SELECT \
  trow,\
  GROUP_CONCAT(if(tkey = 'Name', tvalue, NULL)) AS 'Name',\
  GROUP_CONCAT(if(tkey = 'Address', tvalue, NULL)) AS 'Address',\
  GROUP_CONCAT(if(tkey = 'Role', tvalue, NULL)) AS 'Designation' \
FROM KeyValueTable \
GROUP BY trow \
) t2 where Name is NOT NULL"      
runSELECT(cmd)
```

## The Views

```
runCMD('drop view if exists v_employee')
cmd ="\
create view v_employee as \
Select t2.Name, t2.Address,t2.Designation from (\
SELECT \
  trow,\
  GROUP_CONCAT(if(tkey = 'Name', tvalue, NULL)) AS 'Name',\
  GROUP_CONCAT(if(tkey = 'Address', tvalue, NULL)) AS 'Address',\
  GROUP_CONCAT(if(tkey = 'Role', tvalue, NULL)) AS 'Designation' \
FROM KeyValueTable \
GROUP BY trow \
) t2 where Name is NOT NULL \
"
runCMD(cmd)
```

```
runSELECT('select * from v_employee')
```

```
runCMD('drop view if exists v_dept1')
cmd = "\
create view v_dept1 as \
Select t2.Department, t2.Location,t2.DepartmentCode from (\
SELECT \
  trow,\
  GROUP_CONCAT(if(tkey = 'DeptName', tvalue, NULL)) AS 'Department',\
  GROUP_CONCAT(if(tkey = 'Location', tvalue, NULL)) AS 'Location',\
  GROUP_CONCAT(if(tkey = 'DeptCode', tvalue, NULL)) AS 'DepartmentCode' \
FROM KeyValueTable \
GROUP BY trow \
) t2 where Department is NOT NULL \
;"
runCMD(cmd)
```

```
runSELECT('select * from v_dept1')
```

```
runCMD('drop view if exists v_dept2')
cmd = "\
create view v_dept2 as \
Select t2.Department, t2.Location,t2.Mgr,t2.DepartmentCode from (\
SELECT \
  trow,\
  GROUP_CONCAT(if(tkey = 'DeptName', tvalue, NULL)) AS 'Department',\
  GROUP_CONCAT(if(tkey = 'Location', tvalue, NULL)) AS 'Location',\
  GROUP_CONCAT(if(tkey = 'Mgr', tvalue, NULL)) AS 'Mgr',\
  GROUP_CONCAT(if(tkey = 'DeptCode', tvalue, NULL)) AS 'DepartmentCode' \
FROM KeyValueTable \
GROUP BY trow \
) t2 where Department is NOT NULL \
;"
runCMD(cmd)
```
